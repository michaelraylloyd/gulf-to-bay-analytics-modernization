```python
import requests

headers = {"User-Agent": "GTB-Portfolio-Demo (michael@example.com)"}
url = "https://api.weather.gov/points/27.9378,-82.7925"

r = requests.get(url, headers=headers)
points = r.json()
points
```

```python
forecast_url = points["properties"]["forecast"]
forecast_url
```

```python
forecast = requests.get(forecast_url, headers=headers).json()
forecast
```

```python
import pandas as pd

# Convert forecast periods into a DataFrame
df = pd.DataFrame(forecast["properties"]["periods"])

# Save locally since pandas cannot write Delta outside Fabric
df.to_csv("noaa_forecast.csv", index=False)

df
```
